```python
import matplotlib.pyplot as plt
import numpy as np

#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)

plt.show()
plt.savefig("filename.png",dpi=300,format="png")
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T=np.sin(X)*np.sin(Y)

plt.scatter(X,Y,alpha=20,s=70,c=T,cmap='coolwarm')

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
